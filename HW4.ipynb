## ข้อ 1.  งบประมาณประจำหรืองบประมาณลงทุน สูงกว่ากัน

```python
#หาผลรวมของงบแต่ละประเภท
annual_budget = sep_oct_df[sep_oct_df['รายการ'] =='ประจำ'][['งบประมาณ']].sum()
capital_budget = sep_oct_df[sep_oct_df['รายการ'] =='ลงทุน'][['งบประมาณ']].sum()

#แสดงผลรวมของงบแต่ละประเภท
print(f'งบประมาณประจำ มีมูลค่าเท่ากับ {int(annual_budget):,} บาท')
print(f'งบประมาณลงทุน มีมูลค่าเท่ากับ {int(capital_budget):,} บาท')

print(' ')

budget = int(annual_budget) < int(capital_budget)
cal1 = int(annual_budget) - int(capital_budget) ##เก็บตัวแปรไว้
cal2 = int(capital_budget) - int(annual_budget)

#หาว่างบประเภทใดมีมูลค่าสูงกว่า และสูงกว่าอยู่เท่าใด
if budget is False:
  print(f'ดังนั้น "งบประมาณประจำ" มีมูลค่า สูงกว่า "งบประมาณลงทุน" และสูงกว่าอยู่ {(int(annual_budget) - int(capital_budget)):,} บาท')
else:
  print(f'ดังนั้น "งบประมาณลงทุน" มีมูลค่า สูงกว่า "งบประมาณประจำ" และสูงกว่าอยู่ {(int(capital_budget) - int(annual_budget)):,} บาท')
```

## ข้อ 2. งบรวม เดือน oct หรือ sep อันไหนมากกว่ากัน

```python
#งบรวม เดือนกันยายน (sep-65)
sep_budget = sep_oct_df[sep_oct_df['เดือน-ปี'] =='sep-65'][['งบประมาณ']].sum()

#งบรวม เดือนตุลาคม (oct-65)
oct_budget = sep_oct_df[sep_oct_df['เดือน-ปี'] =='oct-65'][['งบประมาณ']].sum()

print(f'งบรวมในเดือนกันยายน มีมูลค่า {int(sep_budget):,} บาท')
print(f'งบรวมในเดือนตุลาคม มีมูลค่า {int(oct_budget):,} บาท')

print(' ')

#หาว่างบรวมในเดือนใดมีมูลค่ามากกว่า และมากกว่าอยู่เท่าใด
if int(sep_budget) > int(oct_budget):
  print(f'ดังนั้น "งบรวมในเดือนกันยายน" มีมูลค่า มากกว่า "งบรวมในเดือนตุลาคม" และมากกว่าอยู่ {(int(sep_budget) - int(oct_budget)):,} บาท')
elif int(sep_budget) < int(oct_budget):
  print(f'ดังนั้น "งบรวมในเดือนตุลาคม" มีมูลค่า มากกว่ากว่า "งบรวมในเดือนกันยายน" และมากกว่าอยู่ {(int(oct_budget) - int(sep_budget)):,} บาท')
```

## ข้อ 3. แก้ค่า missing พร้อมทั้งให้เหตุผล

```python
sep_oct_df = pd.concat([sep_df,oct_df])
sep_oct_df
```

```python
sep_oct_df.rename(columns=lambda x: x.replace('\n', '')) #ลบค่า \n ในหัวข้อคอลัมน์
sep_oct_df.rename(columns=lambda x: x.replace('*/-', '')) #ลบค่า */- ในหัวข้อคอลัมน์

#เปลี่ยนค่า NaN ให้เป็นค่า -
#เนื่องจากไม่มีข้อมูลที่แสดงให้เห็นจากตาราง

sep_oct_df.fillna('-',inplace=True) #หากใช้ .dropna จะทำให้ทุกแถวหายไป เนื่องจากในทุกๆ แถวมีค่า NaN ทั้งหมด
sep_oct_df
```

## ข้อ 4. เดือนไหนเบิกจ่ายรวมมากกว่ากัน (การใช้จ่าย (PO+เบิกจ่าย))

```python
print('การใช้จ่าย (PO+เบิกจ่าย) โดยแบ่งแยกตามรายการในแต่ละเดือน แสดงรายละเอียดดังนี้')
print(' ')

ann_sep_pay = sep_oct_df[(sep_oct_df['รายการ']=='ประจำ') & (sep_oct_df['เดือน-ปี']=='sep-65')]['การใช้จ่าย (PO+เบิกจ่าย)']
print(f'งบประจำในเดือนกันยายน มีการเบิกจ่ายอยู่ที่ {int(ann_sep_pay):,} บาท')

cap_sep_pay = sep_oct_df[(sep_oct_df['รายการ']=='ลงทุน') & (sep_oct_df['เดือน-ปี']=='sep-65')]['การใช้จ่าย (PO+เบิกจ่าย)']
print(f'งบลงทุนในเดือนกันยายน มีการเบิกจ่ายอยู่ที่ {int(cap_sep_pay):,} บาท')

ann_oct_pay = sep_oct_df[(sep_oct_df['รายการ']=='ประจำ') & (sep_oct_df['เดือน-ปี']=='oct-65')]['การใช้จ่าย (PO+เบิกจ่าย)']
print(f'งบประจำในเดือนตุลาคม มีการเบิกจ่ายอยู่ที่ {int(ann_oct_pay):,} บาท')

cap_oct_pay = sep_oct_df[(sep_oct_df['รายการ']=='ลงทุน') & (sep_oct_df['เดือน-ปี']=='oct-65')]['การใช้จ่าย (PO+เบิกจ่าย)']
print(f'งบลงทุนในเดือนตุลาคม มีการเบิกจ่ายอยู่ที่ {int(cap_oct_pay):,} บาท')

print(' ')

#ผลรวมการเบิกจ่ายรายเดือน
sep_pay = sep_oct_df[sep_oct_df['เดือน-ปี'] =='sep-65'][['การใช้จ่าย (PO+เบิกจ่าย)']].sum()
oct_pay = sep_oct_df[sep_oct_df['เดือน-ปี'] =='oct-65'][['การใช้จ่าย (PO+เบิกจ่าย)']].sum()

print('จะได้ว่า')
print(f'ในเดือนกันยายน มีการเบิกจ่ายอยู่ที่ {int(sep_pay):,} บาท')
print(f'ในเดือนตุลาคม มีการเบิกจ่ายอยู่ที่ {int(oct_pay):,} บาท')

print(' ')

#หาว่าเดือนใดมีการเบิกจ่ายรวมมากกว่า และมากกว่าอยู่เท่าใด
pay = int(sep_pay) > int(oct_pay)
if pay is True:
  print(f'ดังนั้น ในเดือน "กันยายน" มีการเบิกจ่ายมากกว่า "ตุลาคม" และมากกว่าอยู่ {(int(annual_budget) - int(capital_budget)):,} บาท')
else:
  print(f'ดังนั้น ในเดือน "ตุลาคม" มีการเบิกจ่ายมากกว่า "กันยายน" และมากกว่าอยู่ {(int(capital_budget) - int(annual_budget)):,} บาท')
```
